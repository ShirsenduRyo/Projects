```python
import pandas as pd
import json

# Load the JSON file
with open('Data/cancer_1000.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Use pandas.json_normalize to flatten the JSON data
df = pd.json_normalize(data)

# Rename the columns to have underscores
df.columns = [col.replace('.', '_') for col in df.columns]

# Display the resulting DataFrame
print(df.shape)
df.head()
```
